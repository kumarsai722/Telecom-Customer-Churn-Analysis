# check outliers

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['age'] < lower_bound) |
              (df['age'] > upper_bound)]

print(outliers)
```

```python
print(outliers.shape[0])
```
